```python
def parse_line(line):
    # Step 1: Strip the line to remove leading/trailing whitespace
    stripped_line = line.strip()
    
    # Step 2: Extract the element before the first space for the first column
    first_space_index = stripped_line.find(' ')
    first_col = stripped_line[:first_space_index]
    
    # Step 3: Extract the element after the last space for the last column
    last_space_index = stripped_line.rfind(' ')
    last_col = stripped_line[last_space_index+1:]
    
    # Step 4: Strip the text again to remove the extracted parts
    remaining_text = stripped_line[first_space_index+1:last_space_index].strip()
    
    # Step 5: Extract the element after the last space in the remaining text for the third column
    second_last_space_index = remaining_text.rfind(' ')
    third_col = remaining_text[second_last_space_index+1:]
    
    # Step 6: Strip again and save the remaining string for the second column
    second_col = remaining_text[:second_last_space_index].strip()
    

    
    return (first_col, second_col, third_col, last_col)
```

```python
def process_text_file(file_path):
    data = []
    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            lines = file.readlines()
            for line in lines:
                line = line.strip()
                if line:  # Check if the line is not empty
                    parsed_line = parse_line(line)
                    data.append(parsed_line)
            return data
    except UnicodeDecodeError as e:
        print(f"Error reading {file_path}: {e}")
    df=pd.DataFrame(data)
```

```python
# rename the dataframe
df=join_df


# Convert 'Year of NZ Schools' to integer for proper sorting
df['Year of NZ Schools'] = df['Year of NZ Schools'].astype(int)

# Sort by 'Year of NZ Schools'
df = df.sort_values(by='Year of NZ Schools')

# Group by 'Institution Number' and maintain the order of 'Year of NZ Schools'
grouped = df.groupby('Institution Number')
```

# Make yes and no consistent in spelling

```python
# let move on to the columns of if Adjacent School with Enrolment Scheme Exists, due to the text processing reason, some spelling of yes and no are not in consistent, so we need to fix this before dropping any duplicates. 

df['Adjacent School with Enrolment Scheme Exists'].unique()
```

```python
# Replace all variations of 'yes' and 'no'
df['Adjacent School with Enrolment Scheme Exists'] = df['Adjacent School with Enrolment Scheme Exists'].replace(['yes', 'Yes', 'Ye', 'YEs', 'YES'], 'Yes')
df['Adjacent School with Enrolment Scheme Exists']=df['Adjacent School with Enrolment Scheme Exists'].replace(['no', 'No', 'N', 'NO'], 'no')
```

```python

# Create a subset DataFrame for each group, drop duplicates, and display the groups
subset_list = []

for name, group in grouped:
    group = group.drop_duplicates(subset=['Institution Number', 'Date Enrolment Scheme was Approved', 'Adjacent School with Enrolment Scheme Exists'])
    print(f"Group: {name}")
    print(group)
    print()
    subset_list.append(group)

# Concatenate all the subsets into a single DataFrame if needed
subset_df = pd.concat(subset_list).reset_index(drop=True)

subset_df.shape
```
